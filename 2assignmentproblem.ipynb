```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic.csv")
#Displayonlycolumns:A.Name,Sex,Age,Fare,Survived
print(df[["Name","Sex","Age","Fare","Survived"]])
result = df[(df["Sex"] == "female") & (df["Fare"] > 30)]
print(result)
```

```python
# Group by Pclass and compute statistics
result = df.groupby("Pclass").agg({
    "Survived": "mean",
    "Fare": "mean",
    "Age": "mean"
})

# Rename columns
result.columns = [
    "Survival Rate",
    "Average Fare",
    "Average Age"
]

# Display result
print(result)
```
