```python
import pandas as pd 
import numpy as np
```

```python
#A
df = pd.read_csv("iris.csv")
#First10rows
print(df.head(10))

#Shape&Datatypes
print(df.shape)
print(type(df))

# Mean
print("Mean:\n")
print(df.mean(numeric_only=True))

# Standard Deviation
print("\nStandard Deviation:\n")
print(df.std(numeric_only=True))

# Minimum Values
print("\nMinimum Values:\n")
print(df.min(numeric_only=True))

# Maximum Values
print("\nMaximum Values:\n")
print(df.max(numeric_only=True))

#B
#petal_length>4.5
result = df[(df["petal_length"]>4.5 ) & (df["species"] == "Iris-virginica" )]
print(result)
```

```python
#C
df = pd.read_csv("iris.csv")

# Group by species and compute statistics
result = df.groupby("species").agg({
    "sepal_length": "mean",
    "petal_width": "max",
    "sepal_width": "std"
})

# Rename columns
result.columns = [
    "Average Sepal Length",
    "Maximum Petal Width",
    "Std Dev of Sepal Width"
]
print(result)
```

```python

df = pd.read_csv("iris.csv")

# Create new column
df["petal_ratio"] = df["petal_length"] / df["petal_width"]

# Average petal_ratio for each species
result = df.groupby("species")["petal_ratio"].mean()

# Display result
print(result)
```
